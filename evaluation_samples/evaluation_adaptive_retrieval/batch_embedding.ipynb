```python
from datasets import load_dataset

dataset = load_dataset("Qdrant/dbpedia-entities-openai3-text-embedding-3-large-3072-100K")
```

```python
import pandas as pd

df = dataset["train"].to_pandas()

display(df)
```

```python
import nltk

nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize

from collections import Counter

df["unigrams"] = df["text"].apply(lambda x: word_tokenize(x))

all_unigrams = [unigram for sublist in df['unigrams'] for unigram in sublist]

unigram_counts = Counter(all_unigrams)

# Convert the counts to a DataFrame
unigram_counts_df = pd.DataFrame(unigram_counts.items(), columns=['unigram', 'count'])

# Print the resulting DataFrame
print(unigram_counts_df.sort_values(by="count"))
```

```python
# looking at the DF in data viewer, this cut off seemed reasonable

display(unigram_counts_df[unigram_counts_df["count"] < 12501].sort_values(by="count"))
```
